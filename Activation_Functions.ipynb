```python
def sigmoid(x):
  s=1.0/(1+np.exp(-x))
  ds=s*(1-s)
  return s,ds
```

```python
x=np.arange(-6,6,0.1)
s_values,ds_values=sigmoid(x)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,s_values,label="sigmoid_function")
plt.plot(x,ds_values,label="derivative of sigmoid")
plt.title("sigmoid fucntion nad its derivative")
plt.xlabel("x")
plt.ylabel("value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def tanh(x):
  t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
  dt=1-t**2
  return t,dt

```

```python
#x=np.array([1,2,3,4,5])
x=np.linspace(-6,6,100)
tanh_values,dt_values=tanh(x)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,tanh_values,label="tanh_function")
plt.plot(x,dt_values,label="derivative of tanh")
plt.title("tanh function and its derivative")
plt.xlabel("x")
plt.ylabel("value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def relu(x):
  return np.maximum(0,x)
```

```python
x=np.linspace(-6,6,100)
relu_values=relu(x)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,relu_values,label="relu_function")

plt.title("Relu Activation function ")
plt.xlabel("x")
plt.ylabel("relu(x)")
plt.legend()
plt.grid(True)
plt.show()
```
